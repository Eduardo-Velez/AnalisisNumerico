# Simpsons 1/3 mejorado

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# n numero de subsecciones del algoritmo
def simpsons13m(f,a,b,n):
    i = 0
    h = (b-a)/n
    for j in np.arange(n/2):
        i += 2*h*(f(a)+4*f(a+h)+f(a+2*h))/6
        a = a+2*h
    return i
```

```python
f = lambda x: x**3
simpsons13m(f,0,4,4)
```
